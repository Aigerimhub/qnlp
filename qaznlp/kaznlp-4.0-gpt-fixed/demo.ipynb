# 📘 KazNLP Demo
Пример морфологического анализа, генерации и POS-теггинга для казахского языка.

```
from kaznlp import generate_form, generate_verb_form
from kaznlp.analyzer import MorphAnalyzer
```

## 🔤 Морфологический анализ

```
analyzer = MorphAnalyzer()
results = analyzer.analyze("кітаптарымнан")
for res in results:
    print(" → ".join(res))
```

## 🧬 Генерация форм существительных и глаголов

```
print(generate_form("кітап", number="plural", case="ablative"))  # кітаптардан
print(generate_verb_form("бар", tense="future", person="1pl"))  # барамыз
```
